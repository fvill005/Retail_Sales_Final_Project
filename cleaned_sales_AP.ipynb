```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
#import the dataset csv

sales = pd.read_csv('Retail_Sales_Data/Sales.csv')
sales.head()
```

```python
#changing date to pandas datetime format, for ease of use since some entries arent formatted the same.
sales['Date'] = pd.to_datetime(sales['Date'])
sales.head()
sales.shape
```

```python
sales.duplicated(['Date']).sum()
```

```python
sales.drop_duplicates(['Store', 'Date'], keep = 'last')
```

```python
#dropping any duplicates
cleaned_sales = sales.drop_duplicates()
cleaned_sales.head()
```

```python
#renaming date column as to avoid errors in sql
cleaned_sales.rename(inplace= True, columns= {'Date': 'Sale_Date'})
cleaned_sales.head()
```

```python
#saving cleaned data 
```

```python
cleaned_sales.to_csv(r'/tmp/Cleaned_Sales.csv', index = False)
cleaned_sales.to_csv(r'Cleaned_Sales.csv', index = False)
```
